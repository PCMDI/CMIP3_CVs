```python
# import data
with open("230703_cmip3.json") as fH:
    c230703 = json.load(fH)

with open("221017_cmip3.json") as fH:
    c221017 = json.load(fH)

del(fH)
```

```python
# get file lists
c221017nd = []
for count, filePath in enumerate(c221017["!noDateFile"]):
    ind = count+1
    filePath = c221017["!noDateFile"][str(ind)]
    print(ind, filePath)
    print("filePath:", filePath)
    c221017nd.append(filePath)
del(count, ind, filePath)
```

```python
# get file lists
c230703nd = []
for count, filePath in enumerate(c230703["!noDateFile"]):
    ind = count+1
    filePath = c230703["!noDateFile"][str(ind)]
    print(ind, filePath)
    print("filePath:", filePath[0])
    c230703nd.append(filePath[0])
del(count, ind, filePath)
```

```python
# difference the results
type(c221017nd)
type(c230703nd)
diff = list(set(c230703nd).difference(c221017nd))
len(diff)
diff.sort()
diff
```
